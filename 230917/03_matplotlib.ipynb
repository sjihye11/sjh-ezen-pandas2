```python
# 라이브러리 로드
import matplotlib.pyplot as plt
```

```python
# 라인 그래프 그리기
# plot() : 인자값을 하나만 넣는 경우 y축의 데이터가 입력되어 그래프를 출력함.
# 인자 값을 2개 넣는 경우 첫번째 인자값은 x축, 두번째 인자값은 y축으로 데이터가 들어감.
plt.plot([1,2,3,4])
# show() : print()와 흡사
plt.show()
```

```python
plt.plot([1,2,3,4],[4,3,7,30])
plt.show()
```

```python
# 바형 그래프
# bar() : x축 데이터와 y축 데이터가 필수 항목
plt.bar([1,2,3],[80,90,60])
plt.show
```

```python
import numpy as np
```

```python
# 산점도 그래프(넘파이를 불러오세요.)
x = np.random.rand(50)   #50개 값을 랜덤하게 가져오겠다.
y = np.random.rand(50)
print(x)
```

```python
# scatter(): x축 y축 데이터가 필수 항목
plt.scatter(x,y)
plt.show()
```
