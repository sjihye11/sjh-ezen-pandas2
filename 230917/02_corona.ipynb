```python
# 또 다른 방법이 있는데, 이건 '일일사망자'를 가지고 실습해볼게요.
# diff(n)
# 해당하는 컬럼에서 n만큼 이동한 데이터와의 차이를 출력
# n은 1이 기본값
df['총사망자'].diff(1)
```

```python
df['총사망자']-df['총사망자'].shift(1)
# 두 개가 같지요? 더 확실하게 비교해볼까요?
df['총사망자'].diff(1) == df['총사망자']-df['총사망자'].shift(1)
# 첫날은 False가 나오는 이유는 결측치이기 때문이에요. 결측치는 결측치를 가지고 뭘 해도 결측치이기 때문이에요.
```

```python
df['총사망자'].diff(1).fillna(0) == (df['총사망자']-df['총사망자'].shift(1)).fillna(0)
```

```python
# 일일사망자라는 새로운 파생변수를 생성
df['일일사망자'] = df['총사망자'].diff().fillna(0)   #diff() = diff(1)이라서 1굳이 안 적었어요.
```

```python
# apply() : 스리즈 형태의 데이터를 일반함수에 대입하여 새로운 스리즈 형태의 데이터를 생성
# Series class 안에 있는 함수
def change(x):
    # x의 값이 0이라면 '월'을 리턴
    # 1이라면 '화'를 리턴
    # ... 6이라면 '수'를 리턴
    if x==0:
        result='월'
    elif x==1:
        result='화'
    elif x==2:
        result='수'
    elif x==3:
        result='목'
    elif x==4:
        result='금'
    elif x==5:
        result='토'
    else:
        result='일'
    
    return result
```

```python
# 참고해봐.
class Test:
    # 생성자 함수
    def __init__(self, _values, _index = 0):  # 좀 더 series의 형태랑 비슷하게. series는 index설정 안 하면 index는 0부터 그 길이만큼이니까.
                                              # 여기서 _index는 입력 안 하면 기본값 0
        # self.변수명 : 독립적인 변수
        if _index == 0:  #index가 0일 때만 이렇게 한다. 다를 때는 걍 됨.
            self.index = list(range(len(_values)))
        else:
            self.index = _index
        self.values = _values
        
    # 일반 함수
    def apply(self, _func):  #그냥 함수의 이름을 받아오겠다.
        result = list(map(_func, self.values))  #리스트의 형태로 받아올거다. map(함수를 받아서, values의 값[5,4,3,2,1]을 써주겠다.)
        return result
```
