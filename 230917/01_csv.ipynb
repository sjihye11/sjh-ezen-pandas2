### 데이터 파일을 pandas를 이용하여 로드
- read_csv({파일경로}, encoding={인코딩 엔진})
    - 파일의 경로
        - 절대 경로
            - 절대적인 주소
            - 환경이 변하더라도 같은 위치를 지정
            - ex) c:/users/admin/document/a.txt -> 일을 하다가 다른 컴퓨터로 옮겨도, 저 경로를 찾아서 감.
            - ex) 인터넷 url주소
        - 상대 경로(일반적으로 상대 경로를 더 많이 사용하게 될 거얌)
            - 상대적인 주소
            - 환경이 변하면 경로도 변경
            - ./ : 현재 파일의 경로
            - ../ : 현재 경로에서 상위로 이동
            - 폴더명/ : 해당하는 하위 폴더로 이동
            - ex) ../csv/corona.csv(상위폴더로 이동 -> csv 하위 폴더로 이동 -> corona.csv)
    - encoding
        - UTF-8(기본값)
            - 가장 일반적인 인코딩 형태(모든 언어에서)
            - 한글을 로드할 때 문제가 발생할 수 있음(웬만하면 한글 사용 안 하는 게 나음)
        - CP949, EUC-KR
            - UTF-8을 이용하여 파일을 로드할 때 에러가 발생한다면 사용하는 인코딩 엔진
            - 한글 데이터를 로드할 때 주로 사용함
- 그 외의 확장자 파일을 로드
    - read_확장자
        - read_excel({파일의 경로})
        - read_json({파일의 경로})
