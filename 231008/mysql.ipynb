# 데이터의 결합
1. Union 결합 (데이터 간의 단순한 결합)
    - 단순하게 행을 결합하거나 열을 결합하는 방법
    - sql에서는
        - table UNION table2
    - pandas에서는
        - concat([df, df2], axis={0|1})  #일반적으로 concat은 행으로 결합함.
2. Join 결합 (특정한 조건의 맞춰서 열을 결합)
    - 특정한 컬럼의 데이터를 기준으로 하여 해당하는 값들이 일치한다면 열을 결합해줌.
    - sql
        - table join table2 on 조건식
    - pandas
        - merge(df, df2, on = 조건식, how = 결합방식)  #단순했던 concat과는 달리 변수 2개 df, df2가 끝임. 그 초과는 한번에 못함


```python
sql = """
    SELECT
    *
    FROM
    tran_1
"""
cursor.execute(sql)
data1 = cursor.fetchall()
df = pd.DataFrame(data1)
df
```

```python
sql = """
    SELECT
    *
    FROM tran_2
"""
cursor.execute(sql)
data2 = cursor.fetchall()
df2 = pd.DataFrame(data2)
df2
```

```python
# sql 쿼리문을 이용하여 데이터를 단순 행 결합
sql = """
    select * from tran_1
    union
    select * from tran_2;
"""
cursor.execute(sql)
data3 = cursor.fetchall()
df3 = pd.DataFrame(data3)
df3
```

```python
# pandas를 이용하여 데이터프레임을 단순 행 결합
# pd.concat([df, df2], axis=0)
# 근데 concat으로 하면 인덱스도 그대로 걍 붙임
pd.concat([df, df2], axis=0, ignore_index=True) #ignore_index=True하면 인덱스 초기화됨.
```

```python
# 인자값으로 쿼리문을 입력하면 결과를 데이터프레임으로 되돌려주는 함수를 만들어보자(좀 더 편하게 살자).
def sql_load(sql):
    cursor.execute(sql)
    data = cursor.fetchall()
    result = pd.DataFrame(data)
    return result
```

```python
sql = """
    select
    *
    from
    emp
"""
emp = sql_load(sql)
emp
```

```python
sql = """
    select
    *
    from
    dept
"""
dept = sql_load(sql)
dept
```

```python
# join 결합 sql
sql = """
    select
    *
    from
    emp
    left join  #left(여기서 왼쪽은 줄바꿈 전의 왼쪽이니까 emp)를 기준으로 join을 하겠다. 
    dept
    on
    emp.DEPTNO = dept.DEPTNO
"""
sql_load(sql)
```

```python
# pandas를 이용하여 join 결합을 해보자 (merge)
pd.merge(emp, dept, on = 'DEPTNO', how = 'left')
# sql의 join과 다른점은 pandas에서는 DEPTNO가 같으므로 한번만 보여줌.
```

```python
# pandas를 이용하여 join 결합을 해보자 (merge)
pd.merge(emp, dept, on = 'DEPTNO', how = 'right') #right도 해볼까염
```

```python
class Mysql:
    # 생성자 함수 : class가 생성될 때 최초로 한번만 실행이 되는 함수를 의미한다. 초기화 함수라고도 부름.
    def __init__(self, _host, _user, _pass, _db, _port):
        self.host = _host
        self.user = _user
        self.password = _pass
        self.db = _db
        self.port = _port
        
    # sql 쿼리문을 입력값으로 받아서 쿼리문을 실행하고 결과값을 되돌려주는 함수
    def sql_load(self, _sql, _value = []):
        # 데이터베이스 서버와의 연결
        _db = pymysql.connect(
            host = self.host,
            user = self.user,
            password = self.password,
            database = self.db,
            port = self.port
        )
        # Cursor 생성
        cursor = _db.cursor(pymysql.cursors.DictCursor)
        
        # sql 쿼리문을 실행
        cursor.execute(_sql,_value)
        
        # _sql 시작이 select로 시작하는 경우
        if _sql.strip().lower().startswith('select'):
            result = cursor.fetchall()
        # 아닌경우
        else:
            _db.commit()
            result = 'Query OK'
            
        # 데이터베이스와의 연결을 종료
        _db.close()
        
        return result
```
