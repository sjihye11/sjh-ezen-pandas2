```python
sql = """
    select
    *
    from
    emp
    where
    DEPTNO = %s or DEPTNO = %s
"""
value = [10, 20]

pd.DataFrame(DB.sql_load(sql, value))
# %s는 이 뒤에 나오는 특정한 값을 넣어준다는 거
# DEPTNO 값이 10이나 20인 것만 보겠다.
```
