```python
# 선생님 풀이
# 라이브러리 로드
import pandas as pd
```

```python
# 1. Sales Records.csv 파일 로드
sales = pd.read_csv('../csv/Sales Records.csv')
sales.head(1)
```

```python
# 2. 특정 컬럼을 제거(한꺼번에)
sales.drop(['Order ID', 'Total Revenue', 'Total Cost', 'Total Profit', 'Sales Channel'], axis = 1, inplace=True)
```

```python
df = sales.copy()
```

```python
# 컬럼의 이름을 변경
# rename()을 사용하여 컬럼의 이름을 변경
df.rename(
    columns = {
        'Region':'대륙',
        'Country':'국가',
        'Item Type':'상품종류',
        'Order Priority' : '우선순위',
        'Order Date': '주문일자',
        'Ship Date':'발송일자',
        'Units Sold':'개수',
        'Unit Price':'단가',
        'Unit Cost':'원가'
        },
)
df.columns=['대륙', '국가', '상품종류', '우선순위', '주문일자', '발송일자', '개수', '단가', '원가']
df.head(1)
```

```python
# 대륙, 국가를 기준으로 오름차순 정렬
df.sort_values(['대륙', '국가'], ascending = True, inplace=True)
```

```python
# 인덱스를 초기화, 기존의 인덱스 제거
df.reset_index(drop=True, inplace=True)
```

```python
# 새로운 파생변수 생성 -> 총단가, 총원가, 총이윤
df['총단가']=df['개수']*df['단가']
df['총원가']=df['개수']*df['원가']
df['총이윤']=df['총단가']-df['총원가']
```

```python
df.head(1)
```

```python
# 결측치 있나 확인
df.info()
```

1. 국가별로 그룹화, 그룹화 연산으로 총이윤의 합계
2. 총이윤을 기준으로 내림차순 정렬
3. 상위 10개의 데이터를 출력
4. South Korea의 순위가 몇 위인가?

```python
# 국가별 그룹화
# 대괄호 안에 대괄호를 하나 더 넣으면, 안쪽 대괄호를 리스트 형태로 쓸 수 있음.
# '국가'를 기준으로 그룹화를 할 것이므로 df['국가']는 앞쪽에 꼭 들어가야함. 
group_data = df[['국가', '총이윤']].groupby(['국가']).sum()
```

```python
group_data.head(1)
# '국가'는 index임. 여기서 column은 '총이윤'밖에 없음.
```

```python
# group_data의 총이윤을 기준으로 내림차순 정렬한다.
group_data.sort_values(['총이윤'], ascending = False, inplace=True)
```

```python
# 상위 10개의 나라를 출력
group_data.head(10)
```

```python
group_data.loc[['South Korea']]
```

```python
group_data.reset_index(drop=False)
# index가 새로 잡히면서 기존의 index였던 '국가'가 column으로 바뀜.
group_data.reset_index(drop=False, inplace=True)
```

```python
group_data.loc[group_data['국가'] == 'South Korea']
# South Korea의 순위는 48위. (index가 0부터 시작이니까)
```

```python
# 그래프 시각화 라이브러리 로드
# matplotlib이라는 라이브러리가 너무 커서 그 안에 있는 pyplot만 가져오겠다는 뜻임.
import matplotlib.pyplot as plt
```

```python
# 바형 그래프의 x축 데이터
x = group_data.index
# 바형 그래프의 y축 데이터
y = group_data['총이윤']

# 한국의 인덱스값 출력
# 그냥 .index만 하면 리스트 형태로 출력될거야. [47]이라고. 근데 나는 딱 47만 필요하거든.
# 그러니까 리스트의 첫번째 데이터가 필요하다고 쓰는거지 .index[0]으로 
korea_index = group_data.loc[ group_data['국가'] == 'South Korea'].index[0]

# 한국의 총이윤 값을 출력(아래 셀 참고)
Korea_data = group_data.loc[ korea_index , '총이윤']

# 그래프 사이즈를 변경
plt.figure(figsize=(16, 8))

bar_list = plt.bar(x, y)

# bar_list에 있는 korea_index위치의 막대를 선택하여 옵션을 변경
bar_list[korea_index].set_color('r')

# 그래프에 주석을 추가
plt.annotate(
    "South Korea : " + str(korea_index + 1), #주석의 문자열
    xytext = (korea_index + 10, Korea_data + 5000000), #주석 문자열의 위치
    xy = (korea_index, Korea_data), #주석이 설명하는 좌표, 화살표가 무엇을 가리킬 것인지
    arrowprops = dict(
        facecolor = 'y',
        shrink = 0.05 #화살표 머리 크기??? 어쨌든 화살표 모양이 다름.
    )
)

plt.show()
```
