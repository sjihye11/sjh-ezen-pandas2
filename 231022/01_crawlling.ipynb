### 통신 방식
- get
    - 가장 일반적으로 사용하는 통신 방식
    - 웹 브라우저에서 주소창에 주소를 입력하고 요청을 보내는 방식
    - 유저가 서버에게 데이터를 보내는 경우 주소 뒤에 ?로 데이터의 시작을 알리고 key=value의 형태로 데이터를 보낸다.
    - 유저가 입력한 데이터가 주소창에 노출이 되므로 보안상의 문제가 발생할 수 있다.
    - 속도는 가장 빠르다.
    - 주소창에 데이터를 보내기 때문에 데이터의 길이가 제한(max 255자)
- post
    - get 방식에서 보안적인 측면을 강화한 데이터 통신 방식
    - 유저가 보내는 데이터를 body라는 부분에 숨겨서 데이터를 보내는 방식
    - 데이터의 길이 제한이 없다.
    - 데이터 노출이 없으므로 보안이 뛰어나다.
    - 웹 브라우저에서 일반적인 주소창에 주소를 입력하는 방식으로 접속이 불가능
    - get 방식에 비해 속도가 느리다.

```python
# 라이브러리 로드
from bs4 import BeautifulSoup as bs
```

```python
# 파싱 작업 (데이터의 형태를 변환 : beautifulsoup 라이브러리에서 사용이 가능한 형태로)
soup = bs(html_data, 'html.parser')
soup
```

```python
a_list = soup.find_all('a')
```

```python
# a태그들의 문자로만 이루어진 배열 데이터를 생성

# find_all()에서는 get_text() 사용 불가능
# 리스트의 각 항목들에서 get_text() 사용 가능
# 반복문을 이용하여 각 항목들을 대입하여 get_text()를 사용하여 문자열을 새로운 리스트에 추가

# 새로운 리스트 변수를 생성
a_result = []

for i in a_list:
    # i : a_list의 각 항목들이 대입
    # print(i.get_text())
    
    a_result.append(i.get_text())

a_result
```

```python
# map( 함수(행동), 자료형데이터 )
list(map(
    lambda x : x.get_text(),
    a_list
))
```

```python
def change(x):
    return x.get_text()

list(map(
    change, a_list
))
```

# 네이버쇼핑에서 데이터 크롤링 하기

```python
# 실습이요~
url = "https://search.shopping.naver.com/search/all?query=%EC%95%84%EC%9D%B4%ED%8C%A8%EB%93%9C&bt=-1&frm=NVSCPRO"

response = requests.get(url)

# response의 html data만 변수에 대입
html_data = response.text

print(html_data)
```

```python
# BeautifulSoup을 이용하여 데이터 파싱
soup = bs(html_data, 'html.parser')
```

```python
div_data = soup.find("div", attrs = {
    'class' : 'basicList_list_basis__uNBZx'
})
```

```python
div_data.find('div', attrs ={
    'class' : 'adProduct_title__amInq'
}).get_text()
```

```python
ad_title_list = div_data.find_all('div', attrs ={
    'class' : 'adProduct_title__amInq'
})
```

```python
title = []
for t in ad_title_list:
    title.append(t.get_text())
    
title
```

```python
ad_price_list = div_data.find_all('span', attrs = {
    'class' : 'price_num__S2p_v'
})
ad_price_list
```

```python
price_list = list(map(
    lambda x : x.get_text(),
    ad_price_list
))

price_list
```

```python
# 리스트의 마직막 항목을 삭제
price_list.pop()
price_list
```

```python
div_data.find_all('img')
```

```python
ad_img_list = div_data.find_all('img')
```

```python
# img 태그 안에 src 속성 값을 출력(이미지 안에는 텍스트가 없음)
# ad_img_list[0]['src']

img_list = list(map(
    lambda x : x['src'],
    ad_img_list
))

img_list
```
